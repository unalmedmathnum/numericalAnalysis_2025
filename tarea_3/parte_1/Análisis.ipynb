# ***Modelo Básico***

***Análisis de resultados***
---

\begin{align}
\frac{dS}{dt} &= \Pi - \beta S Z - \delta S, \\[6pt]
\frac{dZ}{dt} &= \beta S Z + \zeta R - \alpha S Z, \\[6pt]
\frac{dR}{dt} &= \alpha S Z + \delta S - \zeta R.
\end{align}

\begin{align*}
\Pi   &= 0.1,      &&\text{(tasa de natalidad/entrada de susceptibles)}\\[4pt]
\beta &= 0.0055,   &&\text{(tasa de infección)}\\[4pt]
\alpha&= 0.0075,   &&\text{(tasa de eliminación de zombis)}\\[4pt]
\delta&= 0.0001,   &&\text{(tasa de muerte natural)}\\[4pt]
\zeta &= 0.09,     &&\text{(tasa de resurrección)}
\end{align*}

***1) Instante inicial***

Al inicio (t=0), las tasas dominantes son αSZ ≈ 18.75 (entrada a R) y βSZ ≈ 13.75 (pérdida de S). Dado que Π y δS son insignificantes, dS/dt ≈ -13.7. Esto explica la caída abrupta de S y el rápido crecimiento inicial de R.

***2) Pico de R***

R crece por la entrada αSZ, pero no indefinidamente. Su pico ocurre cuando la entrada iguala a la salida: αSZ + δS = ζR. Luego, al caer S, la entrada a R disminuye y la resurrección ζR lo vacía, generando el descenso.

***3) Comportamiento de Z***

Z crece por dos vías: infección (βSZ) y resurrección (ζR). Las eliminaciones (αSZ) lo reducen temporalmente. A largo plazo, Z alcanza una meseta alta (~510) porque S se agota (sin nuevo R) y las resurrecciones mantienen Z, mientras Π aporta nuevos susceptibles que terminan infectados.

***4) Interconexión S→Z y caída de R***

S alimenta a Z (βSZ) y a R (αSZ). Cuando S cae, la entrada a R se reduce. R entonces decrece por la resurrección ζR (su única salida), mostrando una cola larga. Z se nutre primero de βSZ y después de ζR, dominando al final.

***5) Estado final (Z domina)***

S → 0 y R → 0, Z → meseta alta. Causas: Π aporta susceptibles que rápido son infectados; ζ recicla R a Z; y al no haber S, no hay eliminaciones (αSZ ≈ 0). El balance neto favorece la zombificación total.

El crecimiento zombi se frena al agotarse los susceptibles (S < 50). La resurrección (ζ·R) se vuelve la fuente principal de zombis, mientras la eliminación (α·S·Z) sigue siendo el sumidero clave. ***Fase de Estabilización en Alto Nivel (t > 55 hasta t = 100)***

Los zombis se estabilizan en ≈480 (95% de la población total). Los susceptibles están casi extinguidos (S ≈ 0-5), y los nacimientos son convertidos inmediatamente. Las eliminaciones crean retirados que resurgen como zombis, estableciendo un estado cuasi-estacionario donde el sistema cicla entre Z y R con flujos mínimos. El aumento exponencial de zombis multiplica las interacciones S-Z, haciendo que α·S·Z se convierta en la fuerza dominante. Se establece un ciclo de retroalimentación positiva: más R → más ζ·R → más Z → más α·S·Z → más R. El pico ocurre cuando se alcanza el equilibrio instantáneo α·S·Z + δ·S ≈ ζ·R, con valores numéricos de 32.52 vs 32.36. Este equilibrio marca el momento donde la epidemia alcanza su máxima intensidad de interacciones, representando el punto de no retorno en el colapso del sistema.

***Fase de Desaceleración y Caída (t ≈ 21.7 – 50)***

Tras el pico, R inicia un descenso sostenido por el agotamiento crítico de susceptibles (S≈40 en el pico). El término α·S·Z se debilita severamente, mientras ζ·R mantiene su vigor debido a los niveles sustanciales de R. Esto crea un desbalance persistente donde salidas > entradas, haciendo que dR/dt sea negativo. Los retirados se convierten en "combustible" principal para zombis mediante resurrección. La velocidad del descenso está determinada por la tasa de resurrección ζ, que actúa como drenaje constante del compartimento R hacia Z.

***Fase de Casi-Extinción (t > 50)***

La población de retirados decae asintóticamente hacia cero (R≈1.9 al final). Con S→0, los términos α·S·Z y δ·S se vuelven despreciables, dejando ζ·R como único mecanismo activo que drena R hacia Z. El sistema termina dominado por zombis (≈513 individuos), con retirados casi extinguidos y susceptibles virtualmente desaparecidos, representando el equilibrio catastrófico final del modelo. Este estado final refleja la inevitabilidad del apocalipsis zombi bajo estos parámetros, donde la resurrección garantiza la conversión total de la población.

***Método de Euler***

---

<br>

***Fase Inicial (t = 0 a t ≈ 5)***

En los primeros momentos de la simulación, el error comienza en valores relativamente bajos, oscilando entre aproximadamente 0.5 y 0.8. Estas oscilaciones erráticas iniciales, con pequeños picos y valles, son indicativas de la alta sensibilidad del método de Euler durante la fase más dinámica del sistema. En este período, los zombis comienzan a infectar a los susceptibles a un ritmo acelerado, creando rápidos cambios en las derivadas del sistema. El método de Euler, al basarse únicamente en la pendiente instantánea en cada punto, lucha por capturar adecuadamente estas variaciones iniciales, resultando en estas fluctuaciones en el error

***Fase de Crecimiento Explosivo (t ≈ 5 a t ≈ 12)***

Esta fase representa el período más crítico de la epidemia, donde el error del método de Euler experimenta un crecimiento dramático, alcanzando su máximo entre 2 y 3 unidades. Este comportamiento coincide exactamente con el punto de inflexión de la epidemia zombi, cuando la tasa de infección alcanza su valor máximo. La incapacidad del método de Euler para anticipar cambios rápidos se vuelve particularmente evidente aquí. La fórmula fundamental del método, $$y_{n+1} = y_n + h·f(t_n, y_n),$$ que utiliza solo información local en el punto actual, no puede prever la explosión epidémica inminente. Este error de truncamiento local se acumula rápidamente durante esta fase de alta no linealidad, donde las segundas derivadas y derivadas de orden superior del sistema son significativas.

***Fase de Meseta Alta (t ≈ 12 a t ≈ 50)***

Después del pico epidémico, el error entra en una meseta elevada, manteniéndose en el rango de 0.1 a 1 durante un período prolongado. Esta persistencia del error refleja la transición del sistema hacia una fase donde los susceptibles continúan disminuyendo mientras la población de retirados aumenta gradualmente. Aunque la epidemia ha pasado su punto más crítico, las dinámicas poblacionales siguen siendo lo suficientemente complejas como para que el método de Euler mantenga un error significativo. La naturaleza acumulativa del error en métodos de integración numérica significa que los errores cometidos durante la fase explosiva continúan propagándose a través de la solución.

***Fase de Estabilización Final (t ≈ 50 a t = 100):***

Alrededor de t = 50, se observa un descenso brusco en el error, cayendo desde aproximadamente 0.1 hasta 0.01, seguido de una estabilización en este valor más bajo hasta el final de la simulación. Esta mejora sustancial en la precisión coincide con el sistema acercándose a su estado de equilibrio final. Cuando las poblaciones comienzan a cambiar más lentamente y las derivadas se vuelven menores en magnitud, el error de truncamiento del método de Euler se reduce significativamente. En regiones donde la solución es más suave y las variaciones son menos abruptas, el método de primer orden puede proporcionar aproximaciones más aceptables, aunque el error acumulado de fases anteriores nunca desaparece completamente.

***Método de Heun***

---

<br>

***Fase Inicial (t = 0 a t ≈ 5)***

Desde el inicio de la simulación, el método de Heun exhibe un error notablemente más bajo que Euler, comenzando en valores que típicamente no superan 0.1-0.2. Esta ventaja inicial se debe fundamentalmente a la naturaleza de predictor-corrector del algoritmo, que calcula primero una estimación preliminar (predictor) usando el método de Euler y luego la refina (corrector) promediando las pendientes en el punto inicial y el punto predictor. Esta aproximación de dos etapas permite capturar mejor la curvatura inicial del sistema, reduciendo significativamente las oscilaciones erráticas que afectan al método de Euler durante los momentos más dinámicos de la epidemia.

***Fase de Crecimiento Moderado (t ≈ 5 a t ≈ 12)***

Durante el período crítico de explosión epidémica, el error de Heun experimenta un aumento mucho más controlado que el observado en Euler, alcanzando valores máximos que generalmente se mantienen por debajo de 0.3-0.5. Esta notable mejora se atribuye directamente a la capacidad del método para anticipar parcialmente los cambios rápidos mediante su etapa correctora. La fórmula $$y_{n+1} = y_n + (h/2)·[f(t_n, y_n) + f(t_{n+1}, ỹ_{n+1})]$$ incorpora información sobre la pendiente futura, permitiendo una mejor adaptación a las no linealidades del sistema durante la fase de crecimiento exponencial de la infección zombi.

***Fase de Meseta Controlada (t ≈ 12 a t ≈ 50)***

Después del pico epidémico, el error de Heun se estabiliza en una meseta significativamente más baja que la de Euler, generalmente en el rango de 0.05-0.15. Esta fase evidencia la capacidad superior del método para manejar las transiciones suaves del sistema, donde las poblaciones de susceptibles, zombis y removidos se reajustan gradualmente. El esquema de segundo orden logra mantener un error acumulado mucho menor, ya que en cada paso temporal el error de truncamiento local es proporcional a h³, en contraste con el h² del método de Euler, lo que se traduce en una notable reducción del error global durante períodos prolongados.

***Fase de Convergencia Acelerada (t ≈ 50 a t = 100)***

La fase final del método de Heun muestra una convergencia más rápida y estable hacia errores muy bajos, típicamente descendiendo por debajo de 0.01 y manteniéndose en valores cercanos a 0.001-0.005 hasta el final de la simulación. Esta excelente performance en la región de equilibrio del sistema subraya la eficacia del método cuando las derivadas son pequeñas y el comportamiento del sistema es predominantemente lineal. La naturaleza auto-correctora del algoritmo permite que los errores residuales se amortigüen más eficientemente, resultando en una solución numérica que se aproxima mucho más fielmente a la solución de referencia.

***Método de RK4***

---

<br>

***Fase Inicial de Precisión Inmediata (t = 0 a t ≈ 8)***

Desde los primeros instantes de la simulación, RK4 establece inmediatamente su superioridad numérica con errores que raramente superan 0.01-0.02, aproximadamente dos órdenes de magnitud por debajo del método de Euler y uno por debajo de Heun. Esta precisión inicial extraordinaria se debe a la arquitectura de cuatro etapas del algoritmo, que evalúa cuidadosamente las pendientes en el punto inicial, dos puntos intermedios y el punto final del intervalo. La combinación ponderada $$y_{n+1} = y_n + (h/6)·(k₁ + 2k₂ + 2k₃ + k₄)$$ permite capturar con notable fidelidad la curvatura y las no linealidades del sistema desde el primer paso, prácticamente eliminando las oscilaciones erráticas que afectan a los métodos de orden inferior durante la fase de establecimiento inicial de la epidemia.

***Fase de Resiliencia ante la Explosión Epidémica (t ≈ 8 a t ≈ 20)***

Durante el período de máxima dinámica del sistema, cuando la epidemia zombi alcanza su punto de inflexión más crítico, RK4 exhibe una resiliencia notable con incrementos de error mínimos que generalmente no exceden 0.03-0.05. Esta estabilidad bajo presión se explica por la capacidad del método para aproximar efectivamente los términos de orden superior en la expansión de Taylor de la solución. Mientras Euler y Heun luchan por capturar la aceleración epidémica, RK4 incorpora información suficiente sobre las derivadas de tercer y cuarto orden del sistema, permitiéndole anticipar y adaptarse a los cambios más abruptos en la dinámica poblacional sin comprometer significativamente la precisión.

***Fase de Decaimiento Acelerado del Error (t ≈ 20 a t ≈ 40)***

Posteriormente al pico epidémico, RK4 demuestra una capacidad excepcional para reducir rápidamente su error, descendiendo a valores típicamente inferiores a 0.005-0.01 en un intervalo notablemente más corto que los métodos de orden inferior. Esta fase evidencia la propiedad de auto-corrección inherente a los métodos de alto orden, donde los errores residuales no solo no se acumulan, sino que se disipan activamente a medida que el sistema se estabiliza. La precisión de cuarto orden (error local ∝ h⁵, error global ∝ h⁴) permite que incluso pequeños pasos temporales produzcan mejoras dramáticas en la exactitud de la solución.

# ***Modelo con Cuarentena***

***Análisis de Resultados***
---


\begin{align}
\frac{dS}{dt} &= \Pi - \beta S Z - \delta S,\\[6pt]
\frac{dI}{dt} &= \beta S Z - \rho I - \delta I - \kappa I,\\[6pt]
\frac{dZ}{dt} &= \rho I + \zeta R - \alpha S Z - \sigma Z,\\[6pt]
\frac{dR}{dt} &= \delta S + \delta I + \alpha S Z - \zeta R + \psi Q,\\[6pt]
\frac{dQ}{dt} &= \kappa I + \sigma Z - \psi Q.
\end{align}


\begin{align*}
\Pi   &= 0.1,      &&\text{(tasa de natalidad / entrada de susceptibles)},\\[4pt]
\beta &= 0.0095,   &&\text{(tasa de infección S$\to$ I por contacto con zombis)},\\[4pt]
\delta&= 0.0001,   &&\text{(tasa de muerte natural)},\\[4pt]
\rho  &= 0.005,    &&\text{(tasa de progresión de infectados I a zombis Z)},\\[4pt]
\kappa&= 0.05,     &&\text{(tasa de cuarentena para infectados I)},\\[4pt]
\zeta &= 0.0001,   &&\text{(tasa de resurrección R$\to$ Z)},\\[4pt]
\alpha&= 0.005,    &&\text{(tasa de eliminación de zombis por humanos)},\\[4pt]
\sigma&= 0.02,     &&\text{(tasa de cuarentena para zombis Z)},\\[4pt]
\gamma&= 0.01.     &&\text{(tasa de muerte/ejecución en cuarentena Q$\to$ R)}
\end{align*}

<br>

$$ S(0)=20, I(0)=5, Z(0)=2, R(0)=0, Q(0)=0$$

***1) Población de susceptibles S(t) — Curva roja***

La población susceptible inicia en 20 y experimenta un leve descenso inicial hasta aproximadamente t=10-15, para luego recuperarse de forma sostenida hasta alcanzar un valor cercano a 22 al final de la simulación. Esta caída inicial es causada por el término de infección β S Z, que convierte susceptibles en infectados. Sin embargo, gracias a la implementación de la cuarentena, la población de zombis Z es rápidamente controlada (por aislamiento σ Z y eliminación α S Z), lo que hace que el término β S Z pierda fuerza rápidamente. Una vez la infección se aplana, la dinámica de S pasa a estar dominada por el balance entre el reclutamiento (Π) y la muerte natural (δ S), acercándose lentamente a su equilibrio teórico Π/δ = 1000, aunque en el horizonte simulado solo logra una recuperación parcial. Esto ilustra cómo la cuarentena frena la transmisión, permitiendo que la población susceptible se recupere progresivamente.


***2) Población infectada I(t) — Curva celeste (turquesa)***

Los infectados comienzan en 5 y disminuyen de manera constante y monótona hasta casi desaparecer. Esta rápida desaparición se explica porque los individuos en este compartimento son aislados en cuarentena (κ I = 0.05 I) a un ritmo diez veces mayor que su progresión a zombis (ρ I = 0.005 I). La ecuación gobernante, dI/dt = β S Z - (ρ+δ+κ)I, se ve dominada por el término negativo una vez que Z disminuye y β S Z se vuelve insignificante, resultando en un decaimiento exponencial aproximado de dI/dt ≈ -0.055 I. Biológicamente, esto demuestra que una cuarentena eficaz y temprana sobre los infectados puede suprimir casi por completo este estado intermedio.


***3) Población de zombis Z(t) — Curva verde agua***

Los zombis inician en 2, muestran un leve aumento inicial (hasta ~5) y luego decrecen de forma sostenida hasta aproximadamente 1. El breve crecimiento inicial se debe al flujo desde los infectados (ρ I), que es significativo al principio. No obstante, este flujo se detiene conforme I cae. A partir de entonces, Z se enfrenta a dos pérdidas constantes: la eliminación por humanos (α S Z) y el aislamiento directo en cuarentena (σ Z). La resurrección (ζ R) es demasiado pequeña (ζ=0.0001) para contrarrestar estas pérdidas. En consecuencia, la población zombi es suprimida de manera continua, mostrando que la combinación de cuarentena para infectados y para zombis, junto con la eliminación, es una estrategia de control efectiva.

***4) Población retirada R(t) — Curva azul muy clara***

Los retirados parten de 0, alcanzan un máximo suave alrededor de 3.5 y luego continúan una acumulación lenta hasta ~5. En la fase inicial, el crecimiento se debe principalmente a las eliminaciones de zombis (α S Z) y, en menor medida, a la muerte natural de infectados (δ I). En la fase tardía, aunque α S Z disminuye por el control de Z, surge una nueva fuente importante: la muerte de individuos en cuarentena (γ Q). Dado que la resurrección (ζ R) es mínima, no logra vaciar este compartimento, por lo que R se acumula de manera persistente. Esto refleja que, incluso en un escenario de control exitoso, la cuarentena genera una acumulación moderada y continua de fallecidos.

***5) Población en cuarentena Q(t) — Curva dorada***

Este compartimento, inicialmente en 0, crece rápidamente hasta alrededor de 6 y luego se estabiliza levemente por encima de 5.5. Su dinámica está gobernada por la entrada de infectados (κ I) y zombis (σ Z) aislados, y la salida por muerte (γ Q). Al inicio, la entrada es alta debido al gran número de infectados (κ I = 0.25). Con el tiempo, al caer I y Z, la entrada disminuye y el sistema se aproxima a un equilibrio donde la entrada se iguala a la salida (κ I + σ Z ≈ γ Q). La estabilización de Q indica la formación de un "colchón" poblacional aislado que ayuda a contener el brote pero que persiste en el tiempo.

***Conclusión general del comportamiento***

El modelo SIZRQ con los parámetros de cuarentena describe un escenario de control epidemiológico exitoso. La infección (I) colapsa rápidamente y los zombis (Z) nunca logran establecer un brote sostenido, gracias a la acción combinada del aislamiento de infectados (κ), el aislamiento de zombis (σ) y la eliminación (α). Los susceptibles (S) sufren solo un pequeño impacto y comienzan una recuperación sostenida. Los retirados (R) aumentan de forma moderada y continua, principalmente por muertes en cuarentena. La población en cuarentena (Q) se estabiliza en un nivel bajo, actuando como una barrera persistente contra la reinfección. La resurrección (ζ) es demasiado baja para alterar este resultado.

***Fases temporales y dominio de variables***

Al inicio, las variables que cambian rápidamente (I y Z) dominan la norma del error. Si un método numérico no captura bien sus pendientes pronunciadas, el error aumenta significativamente en esta fase inicial. En tiempos largos, cuando muchas variables tienden a estados estacionarios o valores muy pequeños (Z→0, I→0), los errores absolutos disminuyen, lo que se observa como una reducción en la norma, especialmente evidente en métodos precisos como RK4.

***Múltiples escalas de tiempo en el modelo***

El sistema presenta escalas de tiempo muy diferentes: por ejemplo, la tasa efectiva de eliminación de zombis (~0.12) es rápida, mientras que la progresión de infectados a zombis (ρ=0.005) es mucho más lenta. Los métodos explícitos de bajo orden (como Euler y, en menor medida, Heun) no manejan uniformemente estas diferencias de escala; su error de truncamiento es más pronunciado en las componentes rápidas. En contraste, RK4 captura con alta fidelidad tanto las dinámicas rápidas como las lentas.

***Consecuencia en la gráfica de error***

Esto explica el comportamiento observado: Heun muestra un pico de error inicial cuando las componentes rápidas son dominantes, y luego el error decae a medida que estas se atenúan. RK4, al aproximar correctamente todas las escalas desde el principio, mantiene un error mínimo y estable durante toda la simulación, sin presentar ese pico pronunciado.
